## PCA on expected alpha

```python
import sklearn.decomposition as deco

n_components = 2
x = alphas_exp # expected alphas

pca = deco.PCA(n_components) # n_components is the components number after reduction
x_t = pca.fit(x)
x_r = x_t.transform(x)
print ('explained variance (first %d components): %.2f'%(n_components, sum(pca.explained_variance_ratio_)))##
```

```python
fig = plt.figure()
for i,x in enumerate(x_r[:,0]):
    y = x_r[i,1]
#    plt.text(x,y,"hello")
    if i < 4:
        plt.plot(x,y,'ro')
    else:
        plt.plot(x,y,'bo')
        
fig.show()
#fig.savefig('BvU_PCA.pdf')
```

```python
#sig_up = interesting
sig_up  =  stats.loc[( stats['log2FC'] < 0 )& (stats['TT significant']) ] ["topic"]
sig_down = stats.loc[( stats['log2FC'] > 0 )& stats['TT significant'] ]["topic"] 
```

```python

print pca.components_.shape
loadings = pca.components_
plt.figure()
for i,x in enumerate(loadings[0,:]):
    y = loadings[1,i]
    plt.text(x,y,i)
    if i in sig_up:
        plt.plot(x,y,'ro')
    elif i in sig_down:
        plt.plot(x,y,'go')
    else:
        plt.plot(x,y,'bo')

```

Interesting that the beers are much more alike than the urines( as expected). 

```python

xt = alphas_exp.transpose() # expected alphas

pca2 = deco.PCA(n_components) 
x_r2 = pca2.fit(xt).transform(xt)
print ('explained variance (first %d components): %.2f'%(n_components, sum(pca2.explained_variance_ratio_)))##
```

PCA of alpha on topic. Colours red and green are significant and red is "up regulated " in Urine and green "down-reg"

```python

plt.figure()
for i,x in enumerate(x_r2[:,0]):
    y = x_r2[i,1]
    #plt.text(x,y,i)
    if i in sig_up:
        plt.plot(x,y,'ro')
    elif i in sig_down:
        plt.plot(x,y,'go')
    else:
        plt.plot(x,y,'bo')
```
